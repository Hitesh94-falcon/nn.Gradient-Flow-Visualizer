```python
def count_substring(string,substring):
    start = 0
    end = len(substring)
    count = 0
    while end <= len(string):
        if substring == string[start:end]:
            count += 1 
        start +=1
        end += 1
    return count
count_substring("ABCDBCD","BCD")
```

```python
def swap_case(s):
    k =""
    for char in s:
        if char.isupper() == True:
            j = char.lower()
            k +=j
        if char.islower() == True:
            l=char.upper()
            k+= l
        else:
            k += char
    return k
if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
thickness = int(input()) #This must be an odd number
c = 'H'
```

```python


#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

```python
def print_formatted(number):
    width = len(bin(number))  # width based on the longest binary number

    for i in range(1, number + 1):
        dec = str(i)              # decimal
        octal = oct(i)[2:]        # octal without '0o'
        hexa = hex(i)[2:].upper() # hex without '0x', uppercase
        binary = bin(i)[2:]       # binary without '0b'
        print(f"{dec:>{width}} {octal:>{width}} {hexa:>{width}} {binary:>{width}}")

if __name__ == '__main__':
    n = int(input("Enter a number: "))
    print_formatted(n)
```

```python
size = 3
```

```python
alphabet = [chr(i) for i in range(97,123)]
print(alphabet)
```

```python
alphabet = alphabet[:size]
indices = list(range(size))
indices = indices[:-1] + indices[::-1]
print(indices)
```

```python
for i in indices:
    start_index = i+1
    org =alphabet[-start_index:]
    rev = org[::-1]
    row = rev + org[1:]
    row = "-".join(row)
    width = size*4-3
    row = row.center(width,"-")
    print(row)
```
